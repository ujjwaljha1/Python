**1. Importing matplotlib and some necessary libraries
If we want to plot within an IPython notebook we can use %matplotlib. It comes with two different commands**

%matplotlib notebook for interactive plots
%matplotlib inline for static plots
We will generally use %matplotlib inline

```python
import matplotlib as mb
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x=np.linspace(0,4*np.pi,100)

#Defining wave
sin_y=np.sin(x)
cos_y=np.cos(x)

#Plotting the Wave
plt.plot(x,sin_y,label='Sine')
plt.plot(x,cos_y,label='Cosine')
plt.legend(loc='upper right')
plt.title('Sine wave')
plt.show()
```

Legend is used to add labels to the graph given in the plot function. The loc argument takes an integer or string that specifies the location of the legend on the plot. different values for loc, such as lower right, center, upper right, lower left, etc.
title function adds a title to the plot
'-' will construct a line like in above plots

```python
plt.plot(x, sin_y,  '*')
plt.plot(x, cos_y,  '--')

plt.title('Sine Wave')

plt.show()
```

**Note**

In an IPython notebook the graph will be plotted even without using the plt.show() but in some conditions when we are working on big datasets and want to find some inference we use the plt.show() so it is generally a good practice to use it and also if we are writing a python script then also it's necessary to use the plt.show() command

**2. Setting Styles to the Graph**
2.1 One way to use styles is use plt.style.use('style_name'). List of available styles is:

```python

print(plt.style.available)
```

```python
plt.style.use('grayscale')

plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')

plt.title('Sine & Cosine Wave')

plt.legend()
plt.show()
```


We can use the various available styles. Do check them out

2.2 We can also increase size of the plot
plt.figure(figsize=(length,height))

```python
plt.figure(figsize = (18,5))

plt.style.use('grayscale')

plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')

plt.title('Sine & Cosine Wave')

plt.legend()
plt.show()
```
